## Freedom of Gays

### Age

```python
data.gays_should_live_free.unique()
```

```python
#replacing responses with numbers 
data.gays_should_live_free.replace({"Neither agree nor disagree":"Neutral","Refusal":0,"Don't know":0},inplace = True)
```

```python
#removing responses with dont know and refusal 
df1 = data[data.gays_should_live_free != 0]
```

```python
import seaborn as sns
age = sns.violinplot(y = "age", x = "gays_should_live_free",data = df1, order = ["Agree strongly","Agree","Neutral","Disagree","Disagree strongly"] )
plt.title("How age affects opinions on freedom of gays")
plt.xlabel("Gays should be able to live freely")
```

```python
#other factors, eductaion years, internet usage minutes, how religious, 
```

Those who agree strongly that gays should live freely tend to be from the younger age groups than the older age groups and those that disagree tend disproportionately to be from the 50 - 80 age group. This is expected as older generations tend to be much more conservative due to previous environments of intolerence that they were socialised in. 

### Years of education

```python
education= sns.violinplot(y = "education_years", x = "gays_should_live_free",data = df1, order = ["Agree strongly","Agree","Neutral","Disagree","Disagree strongly"] )
plt.title("How education affects opinions on freedom of gays")
plt.xlabel("Gays should be able to live freely")

```

```python
#remove errors from data where education years are longer than the person has been alive  
df2 = data[(data.age > data.education_years) & (data.education_years >= 0)]
```

```python
education2 = sns.violinplot(y = "education_years", x = "gays_should_live_free",data = df2, order = ["Agree strongly","Agree","Neutral","Disagree","Disagree strongly"] )
plt.title("How education affects opinions on freedom of gays")
plt.xlabel("Gays should be able to live freely")
```

Those who disagree that gays should live freely tend to be disproportionately from those with years of education from 0 - 5 years.
