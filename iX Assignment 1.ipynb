```python
notextnull = notext[notext.salary.notnull()]

notextnull
```

```python
new = notextnull.reset_index()
    
```

```python
#Removing Text

list3 = []

for x in range(0,len(new)):
    if any(char.isalpha() for char in str(new.salary[x])):
        newstr = []
        for char in str(new.salary[x]):
            if char.isalpha():
                pass
            else:
                newstr.append(char)
        word = "".join(newstr)
        list3.append(word)
        
    else:
        list3.append(str(new.salary[x]))
                
list3

newcol = pd.Series(list3)

new["notext"] = newcol
new #new column with salary minus text

#Creating new column with only x - y data
list6 = []

for x in range(0,len(new)):
    list4 = []
    for strings in new.notext[x].split(" "):
        list4.append(strings)
    if "-" in list4:
        list5=[]
        indexdash = list4.index("-")
        numleft = indexdash - 1
        numright = indexdash + 1
        if len(list4) >= numright + 1:
            list5.append(list4[numleft])
            list5.append(list4[indexdash])
            list5.append(list4[numright])
        else:
            list5=[]
        list6.append(list5)
    else:
        list6.append([])
            
list6
newcol2 = pd.Series(list6)

new["format"] = newcol2
new #new column with x - y for those of that format and empty set for those without

```

```python

#Finding average 
list7 = []

for x in range(0,len(new)):
    list22 = []
    list23 = []
    if new.format[x] == []:
        list7.append([])
    else:
        for char in new.format[x][0]:
            if char.isdigit() or char == ".":
                list22.append(char)
            else:
                pass
        num1 = "".join(list22)
        
        for char in new.format[x][2]:
            if char.isdigit() or char == ".":
                list23.append(char)
            else:
                pass
        num2 = "".join(list23)
        
        if num1 == "" or num2 == "":
            list7.append([])
        else:
            ave = (float(num1) + float(num2))/2
            aveint = int(ave)
            list7.append(aveint)
        
newcol3 = pd.Series(list7)
new["ave"] = newcol3
new


#Create final column 
list8 = []

for x in range(0,len(new)):
    if new.ave[x] == []:
        list8.append(new.salary[x])
    else:
        list8.append(new.ave[x])

newcol4 = pd.Series(list8)

new["Final"] = newcol4
new




#Create column to show hours or year
list9 = []
for x in range(0,len(new)):
    if "hour" in str(new.salary[x]):
        list9.append("Hour")
    elif "year" in str(new.salary[x]):
        list9.append("Year")
    else:
        list9.append("NA")

newcol5 = pd.Series(list9)
new["Salary Per"]=newcol5
new

```
